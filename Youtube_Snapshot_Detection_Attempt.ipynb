```python
import shutil
for index, row in youtubevals.iterrows():
#     print(row["Thumbnail URL"])
    image_url=row["Thumbnail URL"]
    filename=row["Thumbnail"]+".jpg"
    r = requests.get(image_url, stream = True)

    # Check if the image was retrieved successfully
    if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
        r.raw.decode_content = True

        # Open a local file with wb ( write binary ) permission.
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)

        print('Image sucessfully Downloaded: ',filename)
    else:
        print('Image Couldn\'t be retreived')
```

```python

from imageai.Classification import ImageClassification
import os

execution_path = os.getcwd()
execution_path
```

```python


prediction = ImageClassification()
prediction.setModelTypeAsDenseNet121()
prediction.setModelPath(os.path.join(execution_path, "DenseNet-BC-121-32.h5"))
prediction.loadModel()

for index,rows in youtubevals.iterrows():
    
    imgname=rows["Thumbnail"]+".jpg"
    predictions, probabilities = prediction.classifyImage(os.path.join(execution_path, imgname), result_count=4 )
#     youtubevals.at[index,'Predictions']=str(predictions)
#     youtubevals.at[index,'Probabilities']=probabilities)
    print("Row",index+1)
    print(predictions, probabilities)
    countiter=1
    for eachPrediction, eachProbability in zip(predictions, probabilities):
        youtubevals.at[index,'Pred-'+str(countiter)]=eachPrediction
        youtubevals.at[index,'Prob-'+str(countiter)]=eachProbability
        countiter+=1
#         print(eachPrediction , " : " , eachProbability)
```

```python
from imageai.Detection import ObjectDetection
import os

execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "yolo.h5"))
detector.loadModel()
for index,rows in youtubevals.iterrows():
    imgname=rows["Thumbnail"]+".jpg"
    imgnameop=rows["Thumbnail"]+"-op.jpg"
    detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path ,imgname), output_image_path=os.path.join(execution_path , imgnameop), minimum_percentage_probability=50)
    countiter=1
    for eachObject in detections:
        print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
        youtubevals.at[index,'AccGuess-'+str(countiter)]=eachObject["name"]
        youtubevals.at[index,'AccPerct-'+str(countiter)]=eachObject["percentage_probability"]
        countiter+=1
        if countiter==6:
            break
        print("--------------------------------")
```

```python
from detoxify import Detoxify
from ProfanityDetector import detector
arr =[]
prof = []
words = []
col = youtubevals.loc[:,"Title"]
for i in range(len(col)):
    word, detected = detector(col[i])
    if detected:
        prof.append(1)
        words.append(word)
    else:
        prof.append(0)
        words.append("None")
    arr.append(col[i])
    

results = Detoxify('original').predict(arr)
newpd = pd.DataFrame(results).round(5)
newpd["profanity"] = prof
newpd["words"] = words
youtubevals = pd.concat([youtubevals,newpd],axis=1, join='inner')
print(youtubevals)
```
